# Documentation

Nous avons appuyé ce projet sur différents articles et travaux de sociologie électorale et de science politique. Nous nous sommes particulièrement intéressés aux travaux sur les déterminants du vote à l'extrême droite (plus spécifiquement du vote pour le Rassemblement National, anciennement Front National).

# Etude de l'électorat de l'extrême gauche

# 1. Un vote de classe ?

Le premier phénomène auquel nous pouvons penser lorsqu'il s'agit du vote à l'extrême gauche est le phénomène des "banlieues rouges", ou de la "ceinture rouge". Ces expressions désignent les villes à mairie communiste entourant Paris depuis les années 1920. Plus largement, ces expressions font le lien entre certaines communes, leur population ouvrière, et leur engagement politique à l'extrême gauche. 

C'est dans ce contexte qu'en 1963, R. Alford développe l'indice qui porte son nom et qui a pour but d'évaluer la proportion d'ouvrier votant à gauche. 

Party and Society (R. Alford, 1963)

# 2. Des enjeux socio-culturels

Plus récemment, en 2012, V. Tiberj postule que l’espace politique français s’articule désormais autour d’une dimension socioéconomique traditionnelle et d’une dimension culturelle. Jusqu’aux années 2000, la droite et la gauche s’opposaient essentiellement sur le terrain socioéconomique mais désormais un second clivage apparaît autour des valeurs culturelles mesurées par des questions relatives aux enjeux de société (homosexualité, place des femmes, PMA…), à l’autoritarisme (peine de mort, discipline à l’école), et à l’ouverture au reste du monde (immigration, perception de la mondialisation.

Cet article justifie donc de s'intéresser aux indicateurs culturels : accès à des théâtres, des cinémas, des musées, etc. Le but de ce projet sera de voir si nos bases de données nous permettent d'établir un lien entre infrastructures et vote des communes lors des élections de 2017.

"La politique des deux axes" (V. Tiberj, 2012)

# 3. Niveau de diplôme

Enfin, d'après T. Piketty, il y a eu une inversion spectaculaire des effets du diplôme. Jusqu'aux années 1980, le vote de droite progresse et le vote de gauche diminue à mesure que s'élèvent les niveaux d'instruction, de revenu et de patrimoine ; les trois courbes sont alignées. A partir des années 1990, le vote de droite continue à croître avec le revenu et la richesse, mais, désormais, c'est le vote de gauche qui croît avec le niveau de diplôme.

Capital et idéologie (T. Piketty, 2019)

# Etude de l'électorat du RN

# 1. Le genre : un déterminant du vote à l'extrême droite ?

On ne peut pas étudier les déterminants du vote pour le RN sans s'intéresser au "Radical Right Gender Gap". Même si ce concept n'est pas étudiable à partir de nos bases de données.

Le Radical Right Gender Gap est l'écart entre les femmes et les hommes dans le vote pour les partis de droite radicale. 

A. Amengy et alii constatent cependant qu'au second tour, il n'y a quasiment pas de différence de genre. Le radical Right Gender Gap semble avoir disparu. Dans cet article, ils expliquent que la précarisation de la situation professionnelle des femmes, en particulier dans le secteur des services, l'érosion de l'effet protecteur du catholiscisme, en lien avec la division de l'Eglise catholique quant à l'attitude à adopter face au FN et surtout l'existence d'un "effet Marine Le Pen". En effet la candidate se présente comme plus moderne, moins clivante, et met en avant son statut de femme divorcée et sa plus grande ouverture en termes de moeurs. Par ailleurs, le programme du parti d'extrême droite insiste sur des thématiques telles que l'égalité salariale et la lutte contre la précarité.


Au-delà d'un effet lié à la prise de contrôle du parti par Marine Le Pen, le recentrage du Front National sur des thématiques économiques peut expliquer en partie le vote féminin. L'électorat féminin du Front National semble en effet plus sensible aux questions relevant de la justice, comme l'expliquent R. Sénac et M. Parodi. 

Globalement, ces articles permettent de mettre en relation les prises de position du parti et ses performances aux élections présidentielles, mais insistent également sur ce que l'on pourrait appeler un "effet candidat". Plus généralement, cela est utile pour savoir qu'il faut prendre du recul sur notre base : étudier l'élection présidentielle ne permet pas de tirer des conclusions générales sur le parti.

"Gender Gap à la française : recomposition ou dépassement ?" (R. Sénac, M. Parodi, 2013)

"L'impact du genre sur le vote Marine Le Pen" (A. Amengy, A. Durovic, N. Mayer, 2017)

# 2. Déterminants socio-économiques

Nous allons pouvoir étudier certains critères d'intégration grâce à nos bases de données, notamment via l'analyse de la présence d'infrastructures publiques dans les communes.

En 2016, S. Crépon et alii montrent dans leur ouvrage qu'il y a une droitisation du vote ouvrier qui correspond à une peur du déclassement. Cette analyse est appuyée l'année suivante par Violaine Girard pour qui si le vote RN ne peut être appréhendé comme un « vote ouvrier », c'est aussi parce que le monde ouvrier apparaît électoralement éclaté. Ainsi, le vote en faveur du RN est nettement plus fréquent au sein des jeunes générations que des plus anciennes : en 2012, les ouvriers ayant entre 18 et 34 ans ont voté deux fois plus souvent pour le FN que ceux de plus de 67 ans. Par ailleurs, cette orientation est davantage le fait des ouvriers stables. Elle mène une enquête de terrain de longue durée (entre 2003 et 2008, puis en 2012) en périphérie de l'agglomération lyonnaise, dans une zone rurale où s'est implanté un parc industriel au début des années 1980 et où le vote FN est élevé. Ce territoire est socialement mixte, mais les ouvriers, majoritairement qualifiés, et les professions intermédiaires de l'industrie (techniciens, contremaîtres) y sont surreprésentés. Elle remet en question l’idée que l'électorat du FN serait composé majoritairement d'ouvriers déçus par la gauche et passés d'un « extrême » à l'autre.



Les faux-semblants du Front National (S. Crépon, A. Dézé, N. Mayer, 2016)

Le vote FN au village. Trajectoires de ménages populaires du périurbain (V. Girard, 2017)
